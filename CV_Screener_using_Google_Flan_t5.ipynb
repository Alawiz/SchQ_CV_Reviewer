```python
# Function to extract text from the PDF file that we will upload
def extract_text_from_pdf(pdf_file):
    pdf_reader = PyPDF2.PdfReader(BytesIO(pdf_file)) # Wrap pdf_file with BytesIO to make it file-like
    text = ""
    for page_num in range(len(pdf_reader.pages)): # Use pdf_reader.pages to get the pages
        page = pdf_reader.pages[page_num] # Access pages using index
        text += page.extract_text()
    return text
```
